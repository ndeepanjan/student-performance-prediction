**Import Necessary Libraries**

```python
import pandas as pd
```

**Create Dataset**

```python
data = {
    "Hours": [1,2,3,3,4,5,6],
    "Internal Marks": [12,14,16,17,18,21,24],
    "Revision": [0,1,1,1.5,2,2,3],
    "Result": ["Fail","Fail","Fail","Pass","Pass","Pass","Pass"]
}

df = pd.DataFrame(data)
print(df)
```

**Convert Result (text → number)**

```python
df["Result"] = df["Result"].map({"Fail":0, "Pass":1})
print(df)
```

**Split input & output**

```python
X = df[["Hours","Internal Marks","Revision"]]
y = df["Result"]
```

**Import ML model**

```python
from sklearn.linear_model import LogisticRegression
```

**Train model**

```python
model = LogisticRegression()
model.fit(X, y)
```

**Prediction**

```python
hours = float(input("Enter study hours (between 1-6): "))
internal = float(input("Enter internal marks (out of 25): "))
revision = float(input("Enter revision hours (between 1-3): "))

pred = model.predict([[hours, internal, revision]])

if pred[0] == 1:
    print("Result: Pass")
else:
    print("Result: Fail")
```
